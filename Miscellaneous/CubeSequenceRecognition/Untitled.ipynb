```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(0, 10, 100)
y = np.zeros(100)
y[:20] = x[:20]
y[20:80] = 2
y[80:] = 10 - x[80:]
```

```python
fig, ax = plt.subplots(1, 1, figsize=(3, 2), dpi=300)
ax.plot(x, y, linewidth = 5)
ax.set_xlabel("x")
ax.set_ylabel("$v_x$")
plt.savefig("traj.png")
```
